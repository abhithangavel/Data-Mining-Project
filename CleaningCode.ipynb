```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import statistics as stats
%matplotlib inline
```

```python
groc = pd.read_csv("groceriesFull.csv")
```

```python
groc.head()
```

```python
columns = groc.columns.tolist()
print(columns)
```

```python
itm_mean = groc['Items'].mean()
print(itm_mean)
```

```python
all_itm = pd.Series(groc.values.ravel('F'))
itm_counts = all_itm.value_counts() 
```

```python
sorted_count = itm_counts.sort_values(ascending = False)
```

```python
#plot
plt.figure(figsize=(10, 6))
sorted_count.plot(kind='bar', color='skyblue', width=3)
plt.title('Common Grocery Items')
plt.xlabel('Grocery Item')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
bars = sorted_count.plot(kind='barh', color='skyblue', width=0.5)

plt.title('Common Grocery Items')
plt.xlabel('Count')
plt.ylabel('Grocery Item')
plt.tight_layout()

plt.show()
```

```python
from wordcloud import WordCloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(all_itm)

plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
